```python
import pandas as pd
```

```python
# Read Data
df = pd.read_csv('us_retail_sales.csv')
```

```python
# Format year to date
df['YEAR'] = pd.to_datetime(df['YEAR'], format='%Y')
```

```python
df = df.set_index('YEAR')
```

```python
df.head()
```

```python
# Area Chart of January
ax = df.plot.area(y='JAN', title='Area Chart Python')
```

```python
# Stacked Area chart all months
ax = df.plot.area(stacked=True, title='Stacked Area Chart Python')
```
