```python
import pandas as pd
```

```python
# Read in Data
data = pd.read_excel('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx.xls')
```

```python
data.head()
```

**Chapter 7**

Fill in Missing Data

```python
# Fills Nan with blank
data_null_filled = data.fillna("BLANK")
```

```python
data_null_filled.head()
```

Remove Duplicates

```python
data_no_dup = data.drop_duplicates()
```

**Chapter 8**

Create hierarchical index

```python
# Creates hierarchical index using Your gender: and How old are you?
data_index = data.set_index(["Your gender:", "How old are you?"])
```

```python
data_index.head()
```

Reshape

```python
# Stacks and reshapes data
data_stack = data_index.stack()
```

```python
data_stack.head(50)
```

**Chapter 10**

Grouping with Functions

```python
# Groups by gender and counts items
data.groupby("Your gender:").count()
```

Grouping with Index Levels

```python
# Groups by age and counts items
data_index_group = data_index.groupby(level="How old are you?").count()
data_index_group.head()
```

**Chapter 11**

Convert between string and date time

```python
# Converts date to string
data['Timestamp'] = data['Timestamp'].astype(str)
data.dtypes
```

```python
# Converts string to date
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.dtypes
```
