```python
import pandas as pd
```

```python
data = pd.read_excel('world-population.xlsm')
```

```python
import matplotlib.pyplot as plt
```

```python
data.head()
```

```python
data.plot(kind='line', x = 'Year', y = 'Population')
plt.title('Line Chart - Python')
plt.show()
```

```python
plt.step(data['Year'], data['Population'], where='pre')
plt.title('Syep Chart - Python')
plt.show()
```
