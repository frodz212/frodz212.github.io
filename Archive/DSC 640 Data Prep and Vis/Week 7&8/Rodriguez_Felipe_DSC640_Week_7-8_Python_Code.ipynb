```python
import pandas as pd
```

```python
df = pd.read_csv('crimerates-by-state-2005.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['murder'], df['robbery'])
plt.xlabel('Murder')
plt.ylabel('Robbery')
plt.title('Scatter Plot Python')
```

```python
import seaborn as sns
```

```python
ax = sns.scatterplot(data=df, x="murder", y="forcible_rape", size="robbery", legend=False, sizes=(20, 2000))
ax.set(title='Bubble Chart Python')
plt.show()
```

```python

sns.distplot(a=df.robbery, kde=False)
plt.title('Density Plot Python')

plt.show()
```
