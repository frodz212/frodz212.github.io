```python
import pandas as pd
```

```python
data = pd.read_excel('hotdog-contest-winners.xlsm')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

Python: Bar Chart

```python
data.plot(kind='bar', x = 'Year', y = 'Dogs eaten')
plt.title('Bar Chart - Python')
plt.show()
```

Python: Stacked Bar Chart

```python
data.plot(kind='bar', x = 'Year', y = ['Dogs eaten', 'New record'], stacked=True)
plt.title('Stacked Bar Chart - Python')
plt.show()
```

Python: Pie Chart

```python
data.groupby(['Country']).sum().plot(kind='pie', y = 'Dogs eaten')
plt.legend(loc='lower right')
plt.title('Pie Chart - Python')
plt.show()
```

Python: Donut Chart

```python
data.groupby(['Country']).sum().plot(kind='pie', y = 'Dogs eaten')
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend(loc='lower right')
plt.title('Donut Chart - Python')
plt.show()
```
